```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
df = pd.read_csv('income.data.csv')
df.head()
```

```python
scatter = [go.Scatter(x = df['income'],y = df['happiness'],mode ='markers')] 
fig = go.Figure(scatter)
iplot(fig)
```

```python
fig = px.imshow(df.corr())
fig.show()
```

```python
hist = [go.Histogram(x = df['income'],\
marker=dict(color ='#AFE400',line = dict(color='black',width=2)))]
fig = go.Figure(data = hist)
iplot(fig)
```

```python
hist = [go.Histogram(x = df['happiness'],\
marker=dict(color ='#AFE400',line = dict(color='black',width=2)))]
fig = go.Figure(data = hist)
iplot(fig)
```

```python
#Assiging values in X & Y
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
#Splitting testdata into X_train,X_test,y_train,y_test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.33,random_state=42)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#Prediciting Y from Linear regression Model
y_pred = lr.predict(X_test)
```

```python
x_range = np.linspace(X.min(), X.max(), 100)
y_range = lr.predict(x_range.reshape(-1, 1))

fig = go.Figure([
        go.Scatter(x=X_train.squeeze(), y=y_train, 
                   name='train', mode='markers'),
        go.Scatter(x=X_test.squeeze(), y=y_test, 
                   name='test', mode='markers'),
        go.Scatter(x=x_range, y=y_range, 
                   name='prediction')
    ])

fig.show()
```

```python
#Assigning Coefficient (slope) to b
b = lr.coef_
```

```python
print("Coefficient  :" , b)
```

```python
#Assigning Y-intercept to a
a = lr.intercept_
```

```python
print("Intercept : ", a)
```

```python
# y_pred=0.71144032(yincome)+0.20878660429859908
print(lr.predict([[7]]))
```

```python
#Mean Squared Error (MSE)
from sklearn import metrics
print('Mean Squared Error (MSE)  : ', metrics.mean_squared_error(y_test, y_pred))
```

```python
import statsmodels.api as sm
X_stat = sm.add_constant(X_train)
regsummary = sm.OLS(y_train, X_stat).fit()
regsummary.summary()
```

```python
print("Adjusted R-Square : ",regsummary.rsquared_adj)
print("R-Square : ",regsummary.rsquared)
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, lr.predict(X_train))
```

```python
plt.scatter(x = y_test, y = lr.predict(X_test))
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```
